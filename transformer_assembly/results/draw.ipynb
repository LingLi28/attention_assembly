```python
import matplotlib.pyplot as plt
import numpy as np

final_data = {
    "Base Stock": {
        "p50": {
            2: {"mean": [10.22, 9.65], "std": [2.71, 0.84]},
            5: {"mean": [10.75, 9.58, 12.38, 9.16, 11.52], "std": [2.87, 0.08, 1.04, 0.65, 1.04]},
            10: {"mean": [11.96, 10.27, 13.07, 11.54, 9.32, 8.46, 11.64, 9.07, 14.19, 14.03],
                 "std": [3.17, 0.32, 0.39, 0.98, 0.29, 0.64, 1.14, 0.70, 1.07, 1.60]},
            20: {"mean": [10.95, 10.21, 11.49, 12.78, 11.58, 14.30, 12.76, 10.67, 12.87, 11.27,
                          15.79, 15.79, 9.69, 12.70, 16.98, 15.15, 13.83, 10.99, 14.22, 12.13],
                 "std": [2.72, 0.29, 0.35, 0.25, 0.10, 0.36, 0.38, 0.52, 0.13, 0.14,
                         0.38, 0.38, 0.47, 0.53, 0.38, 0.13, 0.21, 0.35, 0.77, 1.34]}
        },
        "p100": {
            2: {"mean": [11.26, 9.42], "std": [3.09, 0.83]},
            5: {"mean": [11.24, 9.57, 13.12, 9.85, 12.25], "std": [3.07, 0.08, 1.00, 1.02, 1.01]},
            10: {"mean": [12.39, 10.21, 13.05, 11.54, 9.27, 8.22, 11.25, 8.85, 14.25, 13.51],
                 "std": [3.48, 0.26, 0.34, 0.83, 0.25, 0.57, 0.95, 0.62, 0.87, 1.28]},
            20: {"mean": [10.97, 10.11, 11.38, 12.71, 11.58, 14.20, 12.64, 10.90, 12.85, 11.25,
                          15.67, 15.67, 9.57, 12.62, 16.85, 15.13, 13.77, 10.89, 13.89, 11.46],
                 "std": [2.76, 0.16, 0.19, 0.14, 0.05, 0.19, 0.19, 0.36, 0.05, 0.07,
                         0.19, 0.19, 0.37, 0.36, 0.19, 0.05, 0.09, 0.18, 0.46, 0.98]}
        }
    },
    "REINFORCE": {
        "p50": {
            2: {"mean": [21.01, 7.60], "std": [9.36, 4.74]},
            5: {"mean": [20.82, 9.40, 17.40, 7.70, 15.70], "std": [9.31, 3.67, 6.18, 4.70, 6.85]},
            10: {"mean": [16.98, 10.10, 11.00, 10.10, 9.20, 7.40, 9.20, 7.40, 12.14, 10.10],
                 "std": [6.23, 2.86, 2.88, 2.86, 3.84, 4.80, 3.84, 4.80, 4.12, 2.86]},
            20: {"mean": [13.01, 10.64, 11.44, 12.24, 11.08, 13.42, 12.03, 9.92, 11.67, 10.72,
                          14.18, 14.18, 9.46, 12.34, 14.52, 12.22, 11.69, 9.60, 12.20, 9.95],
                 "std": [3.73, 1.76, 1.82, 1.96, 2.49, 2.42, 1.96, 2.92, 2.61, 2.97,
                         2.97, 2.97, 3.99, 3.77, 3.37, 2.91, 3.24, 4.28, 4.73, 5.17]}
        },
        "p100": {
            2: {"mean": [21.01, 7.60], "std": [9.36, 4.74]},
            5: {"mean": [20.82, 9.40, 17.40, 7.70, 15.70], "std": [9.31, 3.67, 6.18, 4.70, 6.85]},
            10: {"mean": [7.26, 10.37, 13.97, 15.44, 12.44, 8.48, 16.85, 11.18, 21.50, 20.57],
                 "std": [6.70, 2.00, 2.35, 2.80, 2.85, 3.36, 4.12, 3.50, 6.33, 5.78]},
            20: {"mean": [14.15, 10.60, 10.60, 11.40, 10.20, 11.99, 10.60, 9.80, 10.20, 9.80,
                          12.37, 12.37, 8.40, 10.79, 12.54, 10.20, 9.80, 7.90, 10.17, 7.67],
                 "std": [4.09, 1.96, 1.96, 2.01, 2.71, 2.14, 1.96, 3.25, 2.71, 3.25,
                         2.32, 3.32, 4.39, 3.81, 2.47, 2.71, 3.25, 4.62, 4.62, 5.19]}
        }
    },
    "Attention RL": {
        "p50": {
            2: {"mean": [12.81, 16.10], "std": [7.34, 6.51]},
            5: {"mean": [14.57, 7.70, 9.60, 8.98, 11.71], "std": [7.32, 3.36, 3.35, 4.24, 4.37]},
            10: {"mean": [13.49, 16.08, 19.43, 7.09, 8.39, 22.40, 8.39, 22.40, 9.77, 11.59],
                 "std": [6.08, 6.74, 9.11, 3.33, 3.90, 9.36, 3.90, 9.36, 4.13, 3.40]},
            20: {"mean": [11.87, 10.68, 12.28, 13.08, 11.96, 14.85, 13.46, 10.04, 13.14, 11.64,
                          15.99, 15.99, 10.52, 13.898, 16.50, 14.24, 13.58, 11.30, 14.23, 12.23],
                 "std": [3.66, 1.57, 1.78, 2.00, 2.34, 2.91, 2.28, 2.60, 2.76, 3.76,
                         3.87, 3.87, 3.63, 3.93, 4.51, 3.66, 3.63, 4.11, 5.13, 5.44]}
        },
        "p100": {
            2: {"mean": [12.77, 8.90], "std": [7.38, 4.28]},
            5: {"mean": [14.58, 7.70, 15.52, 14.42, 15.97], "std": [7.40, 3.36, 4.60, 5.30, 5.36]},
            10: {"mean": [10.73, 8.69, 6.96, 9.22, 7.11, 8.84, 12.99, 12.44, 18.21, 16.06],
                 "std": [6.54, 2.73, 3.04, 2.55, 3.93, 2.88, 3.14, 3.15, 4.73, 4.95]},
            20: {"mean": [10.73, 10.72, 13.12, 13.92, 12.84, 16.28, 14.89, 10.16, 14.61, 12.56,
                          17.80, 17.80, 11.58, 15.44, 18.48, 16.26, 15.47, 13.00, 16.26, 14.51],
                 "std": [3.94, 1.37, 1.85, 2.13, 2.28, 3.52, 2.80, 2.27, 3.13, 2.62,
                         4.89, 4.89, 3.35, 4.25, 5.76, 4.71, 4.31, 4.14, 5.57, 5.96]}
        }
    }
}



# 
def plot_nodes(file_name, data, p, nodes, title):
    plt.figure(figsize=(12, 6))
    for method, values in data.items():
        x = np.arange(1, len(values[p][nodes]["mean"]) + 1)
        y = np.array(values[p][nodes]["mean"])
        std = np.array(values[p][nodes]["std"])
        plt.plot(x, y, label=method, marker="o")
        plt.fill_between(x, y - std, y + std, alpha=0.2)
    # 
    node_labels = [f"{i}" for i in range(1, len(values[p][nodes]["mean"]) + 1)]
    plt.xticks(ticks=x, labels=node_labels, rotation=0)  # 
    plt.title(f"{title} ({nodes} Nodes, {p})")
    plt.xlabel("Node")
    plt.ylabel("Average Inventory Level")
    plt.legend()
    plt.grid(True, linestyle='--', alpha=0.7)
    plt.tight_layout()
    plt.savefig(file_name, dpi=500)
    plt.show()

def plot_std_combined(file_name, data, nodes, title):
    plt.figure(figsize=(12, 6))
    styles = {"p50": "solid", "p100": "dashed"}  # p
    
    for p in ["p50", "p100"]:
        for method, values in data.items():
            x = np.arange(1, len(values[p][nodes]["std"]) + 1)
            std = np.array(values[p][nodes]["std"])
            
            # 
            plt.plot(x, std, label=f"{method} ({p})", linestyle=styles[p], marker="o")
    
    # 
    node_labels = [f"{i}" for i in range(1, len(values["p50"][nodes]["std"]) + 1)]
    plt.xticks(ticks=x, labels=node_labels, rotation=0)  # 
    
    plt.title(f"{title} ({nodes} Nodes)")
    plt.xlabel("Node")
    plt.ylabel("Standard Deviation")
    plt.legend()
    plt.grid(True, linestyle='--', alpha=0.7)
    plt.tight_layout()
    plt.savefig(file_name, dpi=300)
    plt.show()

```
